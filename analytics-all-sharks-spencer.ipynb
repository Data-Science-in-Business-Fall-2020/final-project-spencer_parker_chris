```python
import pandas as pd
df = pd.read_csv('AllSharkAnalytical.csv')
df.head()
```

```python
df.SharkCount.value_counts()
```

```python
groupFeed = df[df.SharkCount == 12]
notGroup = df[df.SharkCount != 12]

import seaborn as sns
ax = sns.distplot(groupFeed.AvgPiecesPerShark)
ax = sns.distplot(notGroup.AvgPiecesPerShark)
```

```python
aj = sns.distplot(groupFeed.AvgTargetsPerShark)
aj = sns.distplot(notGroup.AvgTargetsPerShark)
```

```python
aj = sns.distplot(groupFeed.AvgDropsPerShark)
aj = sns.distplot(notGroup.AvgDropsPerShark)
```

```python
from scipy import stats
dfstatsgroups = pd.DataFrame(columns=['t-stat', 'p-stat'])
t, p = stats.ttest_ind(groupFeed.AvgPiecesPerShark, notGroup.AvgPiecesPerShark)
dfstatsgroups.loc['AvgPiecesEaten'] = [t, p]

t, p = stats.ttest_ind(groupFeed.AvgTargetsPerShark, notGroup.AvgTargetsPerShark)
dfstatsgroups.loc['AvgTargets'] = [t, p]

t, p = stats.ttest_ind(groupFeed.AvgDropsPerShark, notGroup.AvgDropsPerShark)
dfstatsgroups.loc['AvgDrops'] = [t, p]
```
