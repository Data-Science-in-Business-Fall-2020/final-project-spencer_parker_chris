```python
df = pd.read_csv('All_Shark_Data.csv')
```

```python
df.columns
```

```python
dfSand = df.filter(['Eat_Ross', 'Eat_Chandler', 'SharkCount','Drop_Ross',
       'Drop_Chandler', 'Target_Ross', 'Target_Chandler','Mazuri Vitamins', 'Garlic'])
```

```python
dfSand.dropna(axis=0, inplace=True)
```

```python
dfSand['AvgEat'] = (dfSand.Eat_Ross + dfSand.Eat_Chandler)/2
dfSand['AvgDrop'] = (dfSand.Drop_Ross + dfSand.Drop_Chandler)/2
dfSand['AvgTarget'] = (dfSand.Target_Ross + dfSand.Target_Chandler)/2
```

```python
dfSand.head()
```

```python
dfGray = df.filter(['Eat_GR1', 'Eat_GR2', 'Eat_GR3', 'Eat_GR4','Eat_GR5', 'Drop_GR1', 
        'Drop_GR2', 'Drop_GR3', 'Drop_GR4', 'Drop_GR5', 'Target_GR1', 
        'Target_GR2', 'Target_GR3', 'Target_GR4', 'Target_GR5', 'Mazuri Vitamins', 'Garlic', 
        'SharkCount'])
dfGray.head()
```

```python
dfGray.dropna(axis=0, inplace=True)
dfGray.shape
```

```python
dfGray['AvgEat'] = (dfGray['Eat_GR1'] + dfGray['Eat_GR2'] + dfGray['Eat_GR3'] + dfGray['Eat_GR4'] + dfGray['Eat_GR5']) / 5
```

```python
dfGray['AvgDrop'] = (dfGray['Drop_GR1'] + dfGray['Drop_GR2'] + dfGray['Drop_GR3'] + dfGray['Drop_GR4'] + dfGray['Drop_GR5']) / 5
```

```python
dfGray['AvgTarget'] = (dfGray['Target_GR1'] + dfGray['Target_GR2'] + dfGray['Target_GR3'] + dfGray['Target_GR4'] + dfGray['Target_GR5']) / 5
```

```python
dfGray.head()
```

```python
dfBlackTip = df.filter(['Eat_BT1', 'Eat_BT2', 'Eat_BT3', 'Eat_BT4', 'Eat_BT5', 
        'Drop_BT1', 'Drop_BT2', 'Drop_BT3', 'Drop_BT4', 'Drop_BT5','Target_BT1', 'Target_BT2', 
        'Target_BT3', 'Target_BT4', 'Target_BT5', 'Mazuri Vitamins', 'Garlic',
        'SharkCount'])
dfBlackTip.shape
```

```python
dfBlackTip.isnull().sum()
```

```python
dfBlackTip.dropna(axis=0, inplace=True)
```

```python
dfBlackTip.shape
```

```python
dfBlackTip['AvgEat'] = (dfBlackTip['Eat_BT1'] + dfBlackTip['Eat_BT2'] + dfBlackTip['Eat_BT3'] + dfBlackTip['Eat_BT4'] + dfBlackTip['Eat_BT5']) / 5
```

```python
dfBlackTip['AvgDrop'] = (dfBlackTip['Drop_BT1'] + dfBlackTip['Drop_BT2'] + dfBlackTip['Drop_BT3'] + dfBlackTip['Drop_BT4'] + dfBlackTip['Drop_BT5']) / 5
```

```python
dfBlackTip['AvgTarget'] = (dfBlackTip['Target_BT1'] + dfBlackTip['Target_BT2'] + dfBlackTip['Target_BT3'] + dfBlackTip['Target_BT4'] + dfBlackTip['Target_BT5']) / 5
```

```python
df.loc[df.SharkCount == 12, 'GroupFeed'] = 1
df.loc[df.SharkCount != 12, 'GroupFeed'] = 0
dfAll = pd.DataFrame({'AvgEat': df.AvgPiecesPerShark, 'AvgDrop': df.AvgDropsPerShark, 'AvgTarget': df.AvgTargetsPerShark, 'Garlic': df.Garlic, 'Mazuri Vitamins':df['Mazuri Vitamins'], 'GroupFeed':df.GroupFeed})
dfAll['SharkType'] = 'Combined'
```

```python
dfSand.loc[dfSand.SharkCount == 12, 'GroupFeed'] = 0
dfSand.loc[dfSand.SharkCount != 12, 'GroupFeed'] = 1
dfSand['SharkType'] = 'Sand Bar'
dfBlackTip.loc[dfBlackTip.SharkCount == 12, 'GroupFeed'] = 0
dfBlackTip.loc[dfBlackTip.SharkCount != 12, 'GroupFeed'] = 1
dfBlackTip['SharkType'] = 'Black Tip'
dfGray.loc[dfGray.SharkCount == 12, 'GroupFeed'] = 0
dfGray.loc[dfGray.SharkCount != 12, 'GroupFeed'] = 1
dfGray['SharkType'] = 'Gray Reef'
```

```python
theFrames = [
    dfSand.filter(['SharkType', 'AvgEat', 'AvgDrop', 'AvgTarget', 'Garlic', 'Mazuri Vitamins', 'GroupFeed']), 
    dfBlackTip.filter(['SharkType', 'AvgEat', 'AvgDrop', 'AvgTarget', 'Garlic', 'Mazuri Vitamins', 'GroupFeed']), 
    dfGray.filter(['SharkType', 'AvgEat', 'AvgDrop', 'AvgTarget', 'Garlic', 'Mazuri Vitamins', 'GroupFeed']), dfAll]
```

```python
dfGraph = pd.concat(theFrames)
```

```python
dfGraph.loc[dfGraph.Garlic == 1, 'Garlic'] = 'Garlic'
dfGraph.loc[dfGraph.Garlic == 0, 'Garlic'] = 'No Garlic'
```

```python
dfGraph.loc[dfGraph['Mazuri Vitamins'] ==1, 'Mazuri Vitamins'] = 'Vitamins'
dfGraph.loc[dfGraph['Mazuri Vitamins'] ==0, 'Mazuri Vitamins'] = 'No Vitamins'
```

```python
dfGraph.loc[dfGraph.GroupFeed ==1, 'GroupFeed'] = 'Fed Seperately'
dfGraph.loc[dfGraph.GroupFeed ==0, 'GroupFeed'] = 'Fed Together'
```

## Garlic Graphs

```python
from scipy import stats
label = 'Garlic'
dftTest = pd.DataFrame(columns=['Shark Type', 'Avg with ' + label, 'Avg without ' + label, 't-score', 'p-value', 'Column', 'no'+ label, label, 'total'])
dfBlackTip.name = 'Black Tip'
dfSand.name = 'Sand Bar'
dfGray.name = 'Gray Reef'
dfAll.name = 'Combined'
frames = [dfBlackTip, dfSand, dfGray, dfAll]
for data in frames:
    labelFrame = data[data[label] == 1]
    noFrame = data[data[label] == 0]
    t, p = stats.ttest_ind(labelFrame.AvgEat, noFrame.AvgEat)
    dftTest.loc[data.name + ' Eat'] = [data.name, labelFrame.AvgEat.mean(), noFrame.AvgEat.mean(), t, round(p, 4),'Eat', noFrame.shape[0], labelFrame.shape[0], data.shape[0]]
    t, p = stats.ttest_ind(labelFrame.AvgDrop, noFrame.AvgDrop)
    dftTest.loc[data.name + ' Drop'] = [data.name, labelFrame.AvgDrop.mean(), noFrame.AvgDrop.mean(), t, round(p, 4),'Drop', noFrame.shape[0], labelFrame.shape[0], data.shape[0]]
    t, p = stats.ttest_ind(labelFrame.AvgTarget, noFrame.AvgTarget)
    dftTest.loc[data.name + ' Target'] = [data.name, labelFrame.AvgTarget.mean(), noFrame.AvgTarget.mean(), t, round(p, 4),'Target', noFrame.shape[0], labelFrame.shape[0], data.shape[0]]

dftTest
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
ax = sns.barplot(x='SharkType', 
                 y='AvgEat', 
                 hue='Garlic', 
                 data=dfGraph, 
                 palette="mako_r")
for a in ax.patches:
    ax.annotate(format(a.get_height(), '.2f'), 
                   (a.get_x() + a.get_width() / 2., a.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, -50), 
                   textcoords = 'offset points',
                   fontsize=12,
                   color='white',
                   weight='bold')
ax.set_title('Garlic Pieces Eaten', fontsize=15)
ax.set_ylabel(ylabel='Average Pieces Eaten', fontsize=15)
ax.set_xlabel(xlabel='Shark Type', fontsize=15)
ax.set_xticklabels(labels=['Sand Bar (Significant)', 'Black Tip', 'Gray Reef (Significant)', 'Combined (Significant)'], fontsize=12)
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
ax = sns.barplot(x='SharkType', 
                 y='AvgDrop', 
                 hue='Garlic', 
                 data=dfGraph, 
                 palette="mako_r")
for a in ax.patches:
    ax.annotate(format(a.get_height(), '.2f'), 
                   (a.get_x() + a.get_width() / 2., a.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, -50), 
                   textcoords = 'offset points',
                   fontsize=12,
                   color='white',
                   weight='bold')
ax.set_title('Garlic Drops', fontsize=15)
ax.set_ylabel(ylabel='Average Number of Pieces Dropped', fontsize=15)
ax.set_xlabel(xlabel='Shark Type', fontsize=15)
ax.set_xticklabels(labels=['Sand Bar', 'Black Tip (Significant)', 'Gray Reef', 'Combined (Significant)'], fontsize=12)
plt.show()
```
